```python
def output_alg(word, sentence_id, alg):
  print("##################################")
  print("Using algorithm " + alg.__name__)
  print("Looking for the sense of the word " + word +
                     " in the sentence: " + db[word][sentence_id]['sentence'])
  result = alg(db[word][sentence_id])
  print("Selected definitions:")
  for id in result:
      print(wn.synset(id).definition)
  print("Definition marked as correct: ")
  print(wn.synset(db[word][sentence_id]['correct_synset_id']).definition)
  print("##################################")
  print()
```

```python
algs = [simplified_lesk, parsed_simplified_lesk, simplified_lesk_wordnet, simplified_lesk_wordnet_complete, most_likely_sense]

word = "bancă"
sentence_id = 0

for alg in algs:
  output_alg(word, sentence_id, alg)
```
